```python
import numpy as np
import matplotlib.pyplot as plt

def aver(data):
  return np.sum(data) / np.size(data)
def MNK(x, y):
  k = (aver(x * y) - aver(x) * aver(y)) / (aver(x**2) - aver(x)**2)
  b = aver(y) - k * aver(x)
  return k, b
```

```python
Err = np.array([6.91251, 6.91251, 6.91251, 0.00302467, 0.000144262, 1.05088e-06, 2.2701e-10, 1.35447e-12])
N = np.array([5 * 2**i for i in range(8)])

plt.scatter(np.log(N), np.log(Err))

k, b = MNK(np.log(N[3:]), np.log(Err[3:]))
log_N_range = np.linspace(2.8, 6.7, 10)
plt.plot(log_N_range , k * log_N_range + b, color = 'orange')

print('Slope coefficient:', k)

plt.xlabel('log(N)')
plt.ylabel('log(Err)')

plt.show()
```
