```python
import numpy as np
import matplotlib.pyplot as plt

def aver(data):
  return np.sum(data) / np.size(data)
def MNK(x, y):
  k = (aver(x * y) - aver(x) * aver(y)) / (aver(x**2) - aver(x)**2)
  b = aver(y) - k * aver(x)
  return k, b
```

```python
Err = np.array([8038.52, 509.69, 96.5788, 2.86525, 0.00546459, 0.000311586])
N = np.array([5 * 2**i for i in range(6)])

plt.scatter(np.log(N), np.log(Err))

k, b = MNK(np.log(N), np.log(Err))
log_N_range = np.linspace(1.5, 5.3, 10)
plt.plot(log_N_range , k * log_N_range + b, color = 'orange')

print('Slope coefficient:', k)

plt.xlabel('log(N)')
plt.ylabel('log(Err)')

plt.show()
```
