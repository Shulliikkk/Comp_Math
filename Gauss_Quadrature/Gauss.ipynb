```python
import numpy as np
import matplotlib.pyplot as plt

def aver(data):
  return np.sum(data) / np.size(data)
def MNK(x, y):
  k = (aver(x * y) - aver(x) * aver(y)) / (aver(x**2) - aver(x)**2)
  b = aver(y) - k * aver(x)
  return k, b
```

```python
err = np.array([0.0223289, 2.60404e-05, 2.29691e-08, 1.03335e-08, 9.47235e-09, 9.26426e-09, 9.21256e-09, 9.19966e-09])
h = np.array([10, 5, 2.5, 1.25, 0.625, 0.3125, 0.15625, 0.078125])
log_l_range = np.linspace(0.7, 2.5, 10)
k, b = MNK(np.log(h[:2]), np.log(err[:2]))
print('Slope coefficient:', k)

plt.scatter(np.log(h), np.log(err))
plt.plot(log_l_range, k * log_l_range + b, color = 'orange')
plt.xlabel('log(h)')
plt.ylabel('log(Err)')
plt.title('Зависимость log(Err)(log(h)) для квадратуры Гаусса на N = 5 точках')
plt.show()
```
