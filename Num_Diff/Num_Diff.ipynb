```python
def aver(data):
  return np.sum(data) / np.size(data)
def MNK(x, y):
  n = (aver(x * y) - aver(x) * aver(y)) / (aver(x**2) - aver(x)**2)
  return n
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return np.exp(x)
x_0 = 0
h = np.array([10**(-i) for i in range(16)])
Df_real = np.ones(16)
```

```python
Df_num_4 = (2 * f(x_0) - 5 * f(x_0 + h) + 4 * f(x_0 + 2 * h) - f(x_0 + 3 * h)) / h**2
Eror_4 = abs(Df_num_4 - Df_real)
n = MNK(np.log(h[:4]), np.log(Eror_4[:4]))
print('n =', n)
print('Df_num_4: ', Df_num_4)
print('Eror_4: ', Eror_4)
plt.scatter(np.log(h), np.log(Eror_4))
plt.ylabel('log(Eror)')
plt.xlabel('log(h)')
plt.show()
```

```python
Df_num_5 = (- f(x_0 + 2 * h) + 16 * f(x_0 + h) - 30 * f(x_0) + 16 * f(x_0 - h) - f(x_0 - 2 * h)) / (12 * h**2)
Eror_5 = abs(Df_num_5 - Df_real)
n = MNK(np.log(h[:3]), np.log(Eror_5[:3]))
print('n =', n)
print('Df_num_5: ', Df_num_5)
print('Eror_5: ', Eror_5)
plt.scatter(np.log(h), np.log(Eror_5))
plt.ylabel('log(Eror)')
plt.xlabel('log(h)')
plt.show()
```

```python
Df_num_6 = (15 / 4 * f(x_0) - 77 / 6 * f(x_0 + h) + 107 / 6 * f(x_0 + 2 * h) - 13 * f(x_0 + 3 * h) + 61 / 12 * f(x_0 + 4 * h) - 5 / 6 * f(x_0 + 5 * h)) / h**2
Eror_6 = abs(Df_num_6 - Df_real)
n = MNK(np.log(h[:3]), np.log(Eror_6[:3]))
print('n =', n)
print('Df_num_6: ', Df_num_6)
print('Eror_6: ', Eror_6)
plt.scatter(np.log(h), np.log(Eror_6))
plt.ylabel('log(Eror)')
plt.xlabel('log(h)')
plt.show()
```

```python
plt.scatter(np.log(h), np.log(Eror_4), label = 'N = 3')
plt.scatter(np.log(h), np.log(Eror_5), label = 'N = 4')
plt.scatter(np.log(h), np.log(Eror_6), label = 'N = 5')
plt.ylabel('log(Eror)')
plt.xlabel('log(h)')
plt.legend()
plt.show()
```
