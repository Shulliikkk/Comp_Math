```python
import numpy as np
import matplotlib.pyplot as plt

def aver(data):
  return np.sum(data) / np.size(data)
def MNK(x, y):
  k = (aver(x * y) - aver(x) * aver(y)) / (aver(x**2) - aver(x)**2)
  b = aver(y) - k * aver(x)
  return k, b
```

```python
# N = 3
Err = np.array([0.000631092, 0.000535663, 0.000535663, 0.000379906, 0.00014593, 4.43506e-05])
N = np.array([2 / 2**i for i in range(6)])

plt.scatter(np.log(N), np.log(Err))

k, b = MNK(np.log(N[3:]), np.log(Err[3:]))
log_N_range = np.linspace(-3.2, 0, 10)
plt.plot(log_N_range , k * log_N_range + b, color = 'orange')

print('Slope coefficient:', k)

plt.xlabel('log(N)')
plt.ylabel('log(Err)')
plt.title('N = 3')

plt.show()
```
